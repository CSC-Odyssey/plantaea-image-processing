```python
import os
import cv2
import numpy as np
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import numpy as np
```

# Flipping

```python
# 0 = vertical
# 1 = horizontal
def image_flip(input_folder, output_folder, flip_num, flip_style_string):
    # Loop through all image files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
            image_path = os.path.join(input_folder, filename)
            image = cv2.imread(image_path)

            flipped_image = cv2.flip(image, flip_num)

            # Save
            output_file = os.path.join(output_folder, flip_style_string + "_" + filename)
            cv2.imwrite(output_file, flipped_image)

            print("Flipped image saved to:", output_file)
```

```python
plant_name = "Vitex_nagundo"

# Vertical Flip
image_flip(f"../../images/raws/{plant_name}", f"../../images/aug_Flips/aug_VerticalFlip/{plant_name}", 0, "V_Flipped")
# Horizontal Flip
image_flip(f"../../images/raws/{plant_name}", f"../../images/aug_Flips/aug_HorizontalFlip/{plant_name}", 1, "H_Flipped")

# Horizontal + Vertical Flip
image_flip(f"../../images/aug_Flips/aug_HorizontalFlip/{plant_name}", f"../../images/aug_Flips/aug_HorizontalVerticalFlip/{plant_name}", 0, "V_Flipped")

# Vertical + Horizontal Flip
image_flip(f"../../images/aug_Flips/aug_VerticalFlip/{plant_name}", f"../../images/aug_Flips/aug_VerticalHorizontalFlip/{plant_name}", 1, "H_Flipped")
```

# Rotation

```python
def image_rotate(input_dir, output_dir):
    # Define rotation angle increment and maximum rotation angle
    rotation_angle_increment = 10
    max_rotation_angle = 90

        # Loop through all image files in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            # Read the image
            image_path = os.path.join(input_dir, filename)
            image = cv2.imread(image_path)
    
            # Get image dimensions
            height, width, _ = image.shape
    
            # Calculate padding amount
            padding_amount = int(max(height, width) * 0.4)
    
            # Add padding to the image
            padded_image = cv2.copyMakeBorder(image, padding_amount, padding_amount, padding_amount, padding_amount,
                                              cv2.BORDER_CONSTANT)
    
            # Calculate number of rotations for 1/4 of the image or up to 90 degrees
            num_rotations = min(int(min(padded_image.shape[0], padded_image.shape[1]) / 4 / rotation_angle_increment),
                                int(max_rotation_angle / rotation_angle_increment))
    
            # Loop through multiple rotation angles
            for i in range(0, num_rotations * rotation_angle_increment, rotation_angle_increment):
                # Calculate center of rotation
                center = (padded_image.shape[1] // 2, padded_image.shape[0] // 2)
    
                # Generate rotation matrix
                rotation_matrix = cv2.getRotationMatrix2D(center, i, 1.0)
    
                # Apply rotation to image
                rotated_image = cv2.warpAffine(padded_image, rotation_matrix, (padded_image.shape[1], padded_image.shape[0]))
    
                # Save the rotated image with rotation angle as part of the filename
                output_filename = f"{filename[:-4]}_{i}_rotated.jpg"
                output_path = os.path.join(output_dir, output_filename)
    
                cv2.imwrite(output_path, rotated_image)
```

```python
plant_name = "Vitex_nagundo"

image_rotate(f"../../images/aug_Flips/aug_flip_compiled/{plant_name}", f"../../images/aug_Rotation/{plant_name}")
```
