```python
compiled_lab_1 = pd.read_csv("../../csv/LAB_HOCS/LAB_HOCS_1.csv", index_col=None)
compiled_lab_2 = pd.read_csv("../../csv/LAB_HOCS/LAB_HOCS_2.csv", index_col=None)
compiled_lab_3 = pd.read_csv("../../csv/LAB_HOCS/LAB_HOCS_3.csv", index_col=None)
compiled_lab_4 = pd.read_csv("../../csv/LAB_HOCS/LAB_HOCS_4.csv", index_col=None)
lab = pd.concat([compiled_lab_1, compiled_lab_2, compiled_lab_3, compiled_lab_4])
lab = lab.reset_index(drop=True)
lab
```

```python
import pandas as pd
import ast

# Assuming 'df' is your DataFrame

# Dictionary to store valid row data
valid_rows = {'contours': [],
              'area_hist_values': [],
              'int_hist_values': [],
              'plant': [],}

# Iterate over rows in the DataFrame
for index, row in lab.iterrows():
    try:
        contours = lab['contours'][index]
        area = ast.literal_eval(lab['area_hist_values'][index].replace('.', '.0,').replace('\n', ""))
        int = ast.literal_eval(lab['int_hist_values'][index].replace('.', '.0,').replace('\n', ""))
        plant = lab['plant'][index]
        
        valid_rows['contours'].append(contours)
        valid_rows['area_hist_values'].append(area)
        valid_rows['int_hist_values'].append(int)
        valid_rows['plant'].append(plant)

    except (SyntaxError, ValueError, TypeError):
        # Skip rows that cause evaluation errors
        continue
```

```python
validated_lab = pd.DataFrame(valid_rows)
validated_lab
```

```python
removed_nl = [d.replace('\n', "") for d in validated_lab['contours']]
removed_es = [d.replace('  ', ' ') for d in removed_nl]
added_commas = [d.replace(' ', ',') for d in removed_es]
removed_pre_commas = [d.replace('[,', '[') for d in added_commas]
removed_addi_commas = [d.replace(',,', ',') for d in removed_pre_commas]
cleaned_contours = [ast.literal_eval(d) for d in removed_addi_commas]
```

```python
validated_lab['contours'] = cleaned_contours
```

```python
validated_lab
```

```python
validated_lab.to_csv('../../csv/LAB_HOCS/validated_lab.csv')
```

```python
import cv2
import numpy as np
import os

valid_rows = {'contours': [],
              'area_hist_values': [],
              'int_hist_values': [],
              'plant': [],}

# Create a directory to store the contour images
output_directory = "../../contour_images"
os.makedirs(output_directory, exist_ok=True)

# Iterate over the DataFrame rows
for index, row in validated_lab.iterrows():
    try:
        contour = np.array([validated_lab['contours'][index][0]])  # Replace 'contour' with the actual column name in your DataFrame
    
        plant_name = row['plant']  # Replace 'plant_name' with the actual column name in your DataFrame
        # Create a directory for the specific plant name if it doesn't exist
        plant_directory = os.path.join(output_directory, plant_name)
        os.makedirs(plant_directory, exist_ok=True)
        
        # Generate a blank image with a black background
        image = np.zeros((400, 400, 3), dtype=np.uint8)
        # Create a blank image for drawing contours
        contour_image = np.zeros_like(image)
    
        # Draw the contour on the image
        cv2.drawContours(contour_image, contour, -1, (0, 0, 255), thickness=cv2.FILLED)
    
        contours = validated_lab['contours'][index]
        area = validated_lab['area_hist_values'][index]
        int = validated_lab['int_hist_values'][index]
        plant = validated_lab['plant'][index]
        
        valid_rows['contours'].append(contours)
        valid_rows['area_hist_values'].append(area)
        valid_rows['int_hist_values'].append(int)
        valid_rows['plant'].append(plant)
    
        # Save the image in the plant name folder
        filename = os.path.join(plant_directory, f"{index}.png")  # You can adjust the filename as desired
        cv2.imwrite(filename, contour_image)
    except:
        continue
```

```python
validated_lab2 = pd.DataFrame(valid_rows)
validated_lab2
```

```python
validated_lab2.to_csv('../../csv/LAB_HOCS/validated_lab2.csv')
```
