```python
import os
import cv2
import numpy as np
```

```python
# Folder path containing images
folder_path = '../raws/Vitex_nagundo/'
```

```python
# Define augmentation functions
def rotate(image, angle):
    M = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    return rotated_image
```

```python
def scale(image, factor):
    M = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), 0, factor)
    scaled_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    return scaled_image
```

```python
def flip(image, flip_type):
    if flip_type == 1:
        flipped_image = cv2.flip(image, 1)  # Horizontal flip
    elif flip_type == 2:
        flipped_image = cv2.flip(image, 0)  # Vertical flip
    else:
        flipped_image = image  # No flip
    return flipped_image
```

```python
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Add more image file extensions if needed
        # Load image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Apply image augmentations
        for i in range(5):  # Apply rotation augmentation 5 times
            rotation_angle = np.random.uniform(-45, 45)  # Randomly select rotation angle between -45 and 45 degrees
            rotated_image = rotate(image, rotation_angle)


            augmented_filename = os.path.splitext(filename)[0] + f'_rotate_{i+1}_' + os.path.splitext(filename)[1]  # Add prefix for augmented images
            augmented_image_path = os.path.join(augmented_filename)
            cv2.imwrite(augmented_image_path, rotated_image)

            print(f'Augmented image saved: {augmented_image_path}')

print('Image augmentation complete!')
```
